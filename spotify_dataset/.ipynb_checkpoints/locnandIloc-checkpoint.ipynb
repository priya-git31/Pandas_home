#### ðŸ”¹ 5. Conditional Selection with loc (only)

iloc canâ€™t do conditions, but loc can:
    
1. df.loc[df['artist'] == 'Drake']      # all songs by Drake
2. df.loc[df['year'] >= 2020]           # all songs released in/after 2020
3. df.loc[df['popularity'] > 90, ['title', 'artist', 'popularity']]

```python
df.loc[df['artist']== 'Drake']
```

```python
# all songs released in/after 2020
df.loc[df['year'] >= 2020]
```

```python
# ðŸ”¹ 10 Practice Questions for loc
# Get the first 7 rows of the dataframe (all columns).

df.loc[0:6]
```

```python
# Show only the title column for rows 0â€“10.

df.loc[0:10,'title']
```

```python
# Retrieve the artist and year for the song in row index 5.

df.loc[5,['artist','year']]
```

```python
# Select rows 0â€“4 and columns title, artist, and year.

df.loc[0:4,['artist', 'title', 'year']]
```

```python
# Get all songs where year is 2020 or later.

df.loc[df['year'] >= 2020]
```

```python
# Show the title, artist, and popularity for all songs where popularity > 85.

df.loc[df['popularity']>85,['title', 'popularity','artist']]
```

```python
# Get all songs by a specific artist, for example â€œDrakeâ€.

df.loc[df['artist'] =='Drake']
```

```python
# Select rows [2, 8, 12] and columns title, top genre, and energy.

df.loc[[2,8,12],['title', 'top genre', 'energy']]
```

```python
# Retrieve the last row of the dataframe using its index.
df.loc[df.index[-1]]
```

```python
# Update the popularity of the song â€œBlinding Lightsâ€ to 100.

df.loc[df['title'] == 'Blinding Lights 1', 'popularity'] = 100
df.head(1)
```
