```python
import pandas as pd
import numpy as np
import os

df = pd.read_csv('Top 100 most Streamed - Sheet1.csv')
df.head(2)
```

### 1. duplicated()

Returns a boolean Series â†’ True if a row is a duplicate, False otherwise.

By default, it checks all columns, but you can specify a subset.

First occurrence of a duplicate is False, later ones are True.

```python
df.duplicated()
 # check duplicate rows
```

```python
df['artist'].duplicated()
  # check duplicates only in 'artist'
```

```python
df.duplicated(subset=['artist','title
 # check duplicates based on artist+title                     
```

### 2. drop_duplicates()

Removes duplicate rows.

Keeps the first occurrence by default, but you can keep last or remove all.

```python
df.drop_duplicates()
 # remove duplicate rows
```

```python
# remove duplicate artists (keep first)
df.drop_duplicates(subset=['artist'])
```

```python
df.drop_duplicates(subset=['artist'],keep='last')
# keep last occurrence
```

```python
df.drop_duplicates(subset=['artist'],keep=False)
 # remove all duplicates
```
