```python
import pandas as pd
import numpy as np
import os

df = pd.read_csv('Top 100 most Streamed - Sheet1.csv')
df.head(2)
```

### drop() â†’ Remove Rows or Columns

Removes rows (by index) or columns (by name).

Does not change the DataFrame unless you use inplace=True.

```python
# Drop rows by index
# removes first 2 rows
df.drop([0,1])
```

```python
df.drop(['artist','year'], axis=1)
```

```python
df.drop(5)
```

### isin() â†’ Check Membership in a List

Returns True/False if values exist in a given list.

Used inside loc for filtering.

```python
df['artist'].isin(['Drake'])
```

```python
df.loc[df['artist'].isin(['Drake','The Weeknd'])]
```

### between() â†’ Range Checking

Checks whether values lie within a range (inclusive).

```python
df['year'].between(2018,2019)
```

```python
df.loc[df['year'].between(2019,2020)]
```
